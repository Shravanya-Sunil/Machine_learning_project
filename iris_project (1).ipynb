# cdf curve

it is the cumilative sum of pdf

```python
counts, bin_edges = np.histogram(iris_setosa["petal_length"], bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title("pdf and cdf for setosa")
plt.xlabel("pdf")
plt.ylabel("cdf")
plt.show()
```

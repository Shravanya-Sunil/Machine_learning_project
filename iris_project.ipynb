# conclusion


Petal length and petal width are the 2 parameters to define the 3 species of iris flower.


```python
import seaborn as sns
sns.set_style("whitegrid")
sns.FacetGrid(df , hue = 'variety', height = 5).map(plt.scatter, 'petal_length','petal_width','sepal_length').add_legend()
plt.show()
```

If PL if 0-2 and PW is 0-0.7 it is setosa.
If PL if 3-5.2 and PW is 1-1.7 it is versicolor.
If PL > 5.2 and PW >1.7 it is verginica
